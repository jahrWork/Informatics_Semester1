## Key Uses of Jupyter Notebooks 
The ability to embed explanations, code, and outputs in a single document makes it a great tool for teaching and documenting data science workflows.

Donald Knuth, a famous computer programmer, popularized the idea to combine explanatory plain english text with computer code, which is commonly called literate programming. By adopting this practice, computer programs, as well as other complex information and ideas, could be better explained to a wide range of people.

Notebooks are not just for NASA scientists either: Students, hobbyists, and business people commonly use the fast, interactive workflows to make notebooks that help solve everyday problems.

Examples of math functions: 
$$
   f: x \rightarrow x^2, \ \  \forall x \in \mathbb{R}
$$

$$
   f: n \rightarrow n^2, \ \  \forall n \in \mathbb{Z}
$$

These two math functions can be implemented with the following Python function:

```python
def f(x): 
    return x**3  

y = f(2.)
print( " f(2.) = ", y, type(y) ) 

m = f(2) 
print( " f(2) = ", m, type(m) )
```

## Graphical representations of functions 

Functions can be evaluated in a set of different values to be plotted 

Given a compact $ [a,b] \in \mathbb{R}$, define a partition with $ N+1 $ points
$$
   x_i = a + \frac{(b-a) i}{N}, \ \ \ \ i=0, \ldots N. 
$$
Evaluate the funcion $f(x)$ in these nodes $ x_i $ to obtain $ y_i = f(x_i) $ and plot those points

```python
import matplotlib.pyplot as plt 
import numpy as np 

a, b = 0., 2. 
N = 10 
x = np.zeros(N+1)
for i in range(0,N+1): 
   x[i] = a + (b-a)*i/N 

y = f(x) 
plt.plot(x, y, "." ) # only plot those points 
plt.plot(x, y)       # draw lines between points
plt.show()

```
